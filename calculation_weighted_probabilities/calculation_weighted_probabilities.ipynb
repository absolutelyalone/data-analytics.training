Расчет взвешенных вероятностей

```python
import pandas as pd
```

```python
df = pd.read_csv('denis_1.csv', sep=';', index_col=0)
df
```

```python
dic_age = []
dic_sex = []
for i in range(10):
    df_75 = df.sample(frac=0.75)
    age_df_75 = df_75.groupby(['age_score']).agg({'Target': ['mean']}).reset_index()
    age_df_75.columns = ['age_score', 'mean_Target']
    dic_age.append(age_df_75)
    
    sex_df_75 = df_75.groupby(['sex_score']).agg({'Target': ['mean']}).reset_index()
    sex_df_75.columns = ['sex_score', 'mean_Target']
    dic_sex.append(sex_df_75)
```

```python
df_age = pd.concat(dic_age,axis=0)
df_age
```

```python
df_age_prob = df_age.groupby('age_score')['mean_Target'].mean().rename('age_score_prob').reset_index()
df_age_prob
```

```python
df_sex = pd.concat(dic_sex,axis=0)
df_sex
```

```python
df_sex_prob = df_sex.groupby('sex_score')['mean_Target'].mean().rename('sex_score_prob').reset_index()
df_sex_prob
```

```python
final_df_1 = pd.concat([df_age_prob, df_sex_prob], axis=1)
final_df_1
```

```python
final_df_2 = df.merge(df_age_prob, how='left', on='age_score') \
               .merge(df_sex_prob, how='left', on='sex_score') \
               .sort_values(by="id")
final_df_2
```

```python
dic_total = []
for i in range(10):
    test_75 = final_df_2.groupby('total_score').sample(frac=0.75)
    total_df_75 = test_75.groupby('total_score').agg({'age_score_prob': 'mean', 'sex_score_prob': 'mean'}).reset_index()
    dic_total.append(total_df_75)
```

```python
df_total = pd.concat(dic_total,axis=0)
df_total
```

```python
final_df_3 = df_total.groupby('total_score').agg({'age_score_prob': 'mean', 'sex_score_prob': 'mean'}).reset_index()
final_df_3
```
